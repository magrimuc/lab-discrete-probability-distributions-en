## Bernoulli's Distribution

In a basket containing 100 fruits, there are 60 apples and 40 oranges. 

A fruit is picked randomly from the basket. 

1. **What is the probability that the fruit is apple?**
1. **What is the probability that it is an orange?**

````python
"""
Calculate:
p = probability that the fruit is an apple 
q = probability that the fruit is an orange
"""
appls = 60
orans = 40
tot = appls + orans

p = appls/tot
q = orans/tot

p
````

**Plot the Probability Mass Function (PMF) for the above binomial distribution.**

*Hint: The range of x in a binomial distribution is from `0` to `n`.*

````python
# CDF plot for the number of apples picked (1 to 4)
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

n = 100
p = 0.6
x = np.arange(1, 5)
cdf = stats.binom.cdf(x, n, p)

plt.figure(figsize=(8, 4))
plt.step(x, cdf, where='mid', label='CDF')
plt.scatter(x, cdf, color='red', zorder=3)
plt.xticks(x)
plt.xlabel('Number of apples')
plt.ylabel('Cumulative probability')
plt.title('CDF for picking 1 to 5 apples (p = 0.6)')
plt.grid(alpha=0.3)
plt.legend()
plt.show()
````

## Poisson Distribution

In this challenge you are required to use the Math library in python to solve the problems.

In the UEFA champions league tournament, the average number of goals scored in a match is 2.3. 

**What is the probability that the number of goals scored in a match is 5?**

*Hint: Use the exp() in the math library*

And to calculate the factorial, you can use the following code:

```python
import math
math.factorial(n) # n is the number you want to find the factorial of
```

````python
# Poisson probability for 5 goals when lambda = 2.3
import math

lam = 2.3
k = 5

prob = (math.exp(-lam) * (lam ** k)) / math.factorial(k)
prob_rounded = round(prob, 3)

prob, prob_rounded
````

**Draw a poisson probability distribution plot for the number of goals to be between 0 to 10.**

````python
# Plot Poisson PMF for k = 0 to 10 with lambda = 2.3
import numpy as np
import matplotlib.pyplot as plt
import math

lam = 2.3
x = np.arange(0, 11)
pmf = [math.exp(-lam) * (lam ** k) / math.factorial(k) for k in x]

plt.figure(figsize=(8, 4))
plt.bar(x, pmf, color='skyblue', edgecolor='black')
plt.xticks(x)
plt.xlabel('Number of goals')
plt.ylabel('Probability')
plt.title('Poisson Distribution (lambda = 2.3)')
plt.grid(alpha=0.3)
plt.show()
````
